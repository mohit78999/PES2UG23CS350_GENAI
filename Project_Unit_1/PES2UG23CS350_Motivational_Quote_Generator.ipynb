**Goal**: Generate unique motivational quotes based on keywords like "Success", "Grind", "Peace".

```python
from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
print("Loading GPT-2 model...")
generator = pipeline('text-generation', model='gpt2')
print("Model loaded successfully!\n")
```

```python
keywords = ["Success", "Grind", "Peace"]

print("Keywords for quote generation:")
for i, keyword in enumerate(keywords, 1):
    print(f"{i}. {keyword}")
```

```python
print("="*70)
print("GENERATING MOTIVATIONAL QUOTES")
print("="*70)
print()

for keyword in keywords:

    prompt = f"Motivational quote about {keyword}:"

    result = generator(
        prompt,
        max_new_tokens=30,
        temperature=1.0,
        do_sample=True,
        top_p=0.95,
        num_return_sequences=1
    )

    generated_text = result[0]['generated_text']

    print(f"Keyword: {keyword}")
    print(f"Quote: {generated_text}")
    print("-" * 70)
    print()
```

## Observations

I made this quote generator using GPT-2 model. The idea was to generate motivational quotes for three keywords - Success, Grind and Peace.

I learned the following:-
- The pipeline from transformers library works well, I just gave it the task type and model name
- I set temperature to 1.0 which makes it more random and creative, so every time I run it gives different output
- max_new_tokens=30 keeps the quotes short, otherwise it was generating long paragraphs which didn't look like quotes at all

Results:
The model successfully generated quotes for all three keywords. Each output was different and showed creativity. The quotes were based on the keywords and had unique wording.

Since I used temperature=1.0, each time I run the code it gives different output which makes the quotes unique.
